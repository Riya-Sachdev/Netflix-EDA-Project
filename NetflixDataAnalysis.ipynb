## **Data Cleaning & Handling Missing Values**

```python
cols = ["director", "cast", "country"]
df[cols] = df[cols].fillna("Unknown")
```

```python
df["rating"] = df["rating"].fillna("Not Rated")
```

#### During the data exploration phase , we observed that the **date_added** column is currently stored as an **object** data type . Since this column represents date information , it is more appropriate to convert it to a **datetime** data type

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
```

```python
df["date_added"].isnull().sum()
```

#### After converting `date_added` to datetime, invalid or missing dates became null, so we have to remove these rows to ensure accurate time-based analysis without affecting insights.

```python
df = df.dropna(subset=["date_added"])
```

```python
df.loc[:,"year_added"] = df["date_added"].dt.year
df.loc[:,"month_added"] = df["date_added"].dt.month
```

```python
movies = df[df["type"] == "Movie"].copy()
movies["duration"] = movies["duration"].str.replace("min", "")
movies["duration"] = pd.to_numeric(movies["duration"], errors="coerce")
```

```python
movies["duration"].isnull().sum()
```

```python
movies["duration"] = movies["duration"].fillna(movies["duration"].median())
```

```python
movies["duration"].isnull().sum()
```

## **Temporal Growth Analysis**

```python
year_count = df["year_added"].value_counts().sort_index()

plt.figure(figsize = (8,4))
sbn.lineplot(x=year_count.index , y=year_count.values , marker="o" , color ="green")
plt.title("Year Wise Growth of Netflix Content")
plt.xlabel("Year")
plt.ylabel("Number of Titles Added")
plt.show()
```

### ðŸ“ˆ **Temporal Growth of Netflix Content**

To understand how Netflixâ€™s content library has evolved over time, we analyzed the number of titles added each year.

### ðŸ” **Key Observations**
- Between 2008 and 2014, Netflix added a relatively small number of titles annually, indicating a limited and curated catalog.
- A sharp increase in content additions begins around 2015.
- The number of titles added per year peaks around 2019, reaching nearly 2,000 new titles.
- After 2019, there is a slight decline in annual additions, though the volume remains significantly higher than earlier years.

### ðŸ“Š **Analytical Insight**
This temporal growth trend provides essential context for downstream analyses. Patterns observed in genre popularity, country-wise content production, or content type distribution are strongly influenced by this rapid expansion phase.
